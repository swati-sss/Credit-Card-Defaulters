```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

df = pd.read_csv('/Users/swatisrivastava/Downloads/Credit_Card_Defaulter_prediction/Data_cleaning_and_Exploration/my_cleaned_data.csv')
```

```python
df.head(10)
```

```python
X = df.drop(['ID', 'default payment next month','EDUCATION_CAT', 'others', 'graduate school', 'high school', 'university'], axis=1)
y = df['default payment next month']
```

```python
# Scale continuous features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Convert labels to categorical (if it's a binary classification)
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# Neural network architecture
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(2, activation='softmax')  # Use 2 for binary classification
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, validation_split=0.2, epochs=10, batch_size=32)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {test_accuracy}, Test Loss: {test_loss}')
```

```python
model.save('cnn.h5')
```
